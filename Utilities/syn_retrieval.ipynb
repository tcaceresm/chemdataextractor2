En este notebook tiene como objetivo obtener los sinónimos de los nombres de drogas. Retorna un archivo de texto con todos los sinónimos, separados por el símbolo '$'
Se basa en la herramienta pubchempy (https://github.com/mcs07/PubChemPy)

```python
#Importación del módulo
try:
    import pubchempy as pcp
    print('pubchempy is installed')
except ModuleNotFoundError:
    print('pubchempy is not installed')
```

```python
#La variable drugs_list es una lista que posee los nombres de drogas de las cuales queremos obtener los sinónimos
drugs_list = ['aspirin', 'glucose', 'AZD3965', 'paracetamol']
```

```python
#Obtenemos los ID's correspondientes a las drogas
drugs_ids = [pcp.get_cids(drug)[0] for drug in drugs_list]
```

```python
drugs_ids
```

```python
compounds = [pcp.Compound.from_cid(drug_id) for drug_id in drugs_ids]
```

```python
compounds
```

```python
#En la variable synonyms se encuentran los sinónimos de cada droga. Es una lista de listas, donde cada lista interior corresponde a los sinónimos de una droga
synonyms = [c.synonyms for c in compounds]
```

```python
synonyms
```

```python
#Importamos un pequeño script para guardar los sinónimos en un archivo de texto
import guardar
```

```python
for drug_index in range(0, len(drugs_list)):
    drug_name = drugs_list[drug_index]
    drug_name_syn = synonyms[drug_index]
    drug_name_syn = '$'.join(drug_name_syn)
    guardar.save_syn(drug=drug_name, texto=drug_name_syn)
```
